```python
#Bagof words
dataset = nltk.sent_tokenize(paragraph)
import re
```

```python
dataset
```

```python
for i in range(len(dataset)):
    dataset[i] = dataset[i].lower()
    dataset[i] = re.sub(r'\W',' ', dataset[i])
    dataset[i] = re.sub(r'\s+', ' ',dataset[i])
    print (dataset[i]) # No puctuations, no extra space
```

```python
#To create bag of word we need to create histogram as first step
#word2count will be contain all the unique words mapped with their occurances
```

```python
word2count = {}
for data in dataset:
    words = nltk.word_tokenize(paragraph)
    for word in words:
        if word not in word2count.keys():
            word2count[word] = 1
        word2count[word] += 1
word2count
```

```python
# Selecting best 100 features
import heapq
freq_words = heapq.nlargest(100,word2count,key=word2count.get)
```

```python
freq_words
```

```python
import numpy as np
# Converting sentences to vectors
```

```python
X=[]
for data in dataset:
    vector =[]
    for word in freq_words:
        if word in nltk.word_tokenize(data):
            vector.append(1)
        vector.append(0)
    X.append(vector)
X = np.asarray(X)
```

```python
## IDF Dictionary
```

```python
word_idfs = {}
for word in freq_words:
    #print (word)
    doc_count = 0
    for data in dataset:
        if word in nltk.word_tokenize(data):
            doc_count += 1
   # word_idfs[word] = np.log(len(dataset)/doc_count)
    word_idfs[word] = np.log(len(dataset)/(doc_count+1))
word_idfs
```

```python
#TF matrix
```

```python
# TF Matrix
tf_matrix = {}
for word in freq_words:
    doc_tf = []
    for data in dataset:
        frequency = 0
        for w in nltk.word_tokenize(data):
            if word == w:
                frequency += 1
        tf_word = frequency/len(nltk.word_tokenize(data))
        doc_tf.append(tf_word)
    tf_matrix[word] = doc_tf
```

```python
tf_idfs = {}
for key in word_idfs.keys():
    tf_idfs[key] = np.array(tf_matrix[key]) * word_idfs[key]
```

```python
tf_idfs
```

```python
# Creating the Tf-Idf Model
tfidf_matrix = []
for word in tf_matrix.keys():
    tfidf = []
    for value in tf_matrix[word]:
        score = value * word_idfs[word]
        tfidf.append(score)
    tfidf_matrix.append(tfidf)   
```

```python
def computeTf(wordDict, bow):
    tfDict = {}
    bowCount = len(bow)
    for word,count in wordDict.items():
        tfDict[word] = count/float(bowCount)
    return tfDict
```

```python
tfBowA = computeTf(wordDictA, bowA)
tfBowB = computeTf(wordDictB, bowB)
```

```python
def computeIDF(docList):
    import math
    idfDict = {}
    N = len(docList)
    idfDict = dict.fromkeys(docList[0].keys(),0)
    print (idfDict)
    for doc in docList:
        for word, val in doc.items():
            if val > 0:
                idfDict[word] += 1
    for word,val in idfDict.items():
        idfDict[word]= math.log10(N/ float(val))
    return idfDict
```

```python
idfs = computeIDF([wordDictA, wordDictB])
```

```python
idfs
```

```python
def computeTFIDF(tfBow, idfs):
    tfidf = {}
    for word, val in tfBow.items():
        tfidf[word] = val*idfs[word]
    return tfidf
```

```python
tfidfBowA = computeTFIDF(tfBowA, idfs)
tfidfBowB = computeTFIDF(tfBowB, idfs)
```
